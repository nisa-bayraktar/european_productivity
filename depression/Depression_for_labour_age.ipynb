```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
depage = pd.read_csv('../data/depression_by_age.csv') 
prod = pd.read_csv('../data/productivity.csv')
prod = prod.drop(['Unnamed: 18'],axis=1)
prod = prod.drop(['2020'],axis=1)
depage = depage.drop(['Prevalence - Depressive disorders - Sex: Both - Age: 10 to 14 (Percent)'],axis=1)
depage = depage.drop(['Prevalence - Depressive disorders - Sex: Both - Age: All Ages (Percent)'],axis=1)
depage = depage.drop(['Prevalence - Depressive disorders - Sex: Both - Age: 70+ years (Percent)'],axis=1)
depage = depage.drop(['Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)'],axis=1)
depage = depage.drop(['Prevalence - Depressive disorders - Sex: Both - Age: 15-49 years (Percent)'],axis=1)
depres_age = pd.DataFrame(depage)
productivity = pd.DataFrame(prod)
depres_age['Age:15-69_depression_average']=depres_age.iloc[:,3:7].mean(axis=1,skipna=True)
depression_age= depres_age.pivot_table('Age:15-69_depression_average', ['Entity', 'Code'], 'Year')
#depres_age= depres_age.pivot_table('Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)', ['Entity', 'Code'], 'Year')
#arr = df2['Country'].unique()

#df3 = df1.set_index(['Entity','Code','Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)','Year'],drop=True).unstack('Year')
#df3

depression_age

```

```python
depres_age=depres_age.drop(['Prevalence - Depressive disorders - Sex: Both - Age: 20 to 24 (Percent)'],axis=1)
depres_age=depres_age.drop(['Prevalence - Depressive disorders - Sex: Both - Age: 30 to 34 (Percent)'],axis=1)
depres_age=depres_age.drop(['Prevalence - Depressive disorders - Sex: Both - Age: 15 to 19 (Percent)'],axis=1)
depres_age=depres_age.drop(['Prevalence - Depressive disorders - Sex: Both - Age: 25 to 29 (Percent)'],axis=1)
depres_age=depres_age.drop(['Prevalence - Depressive disorders - Sex: Both - Age: 50-69 years (Percent)'],axis=1)



```

```python
arr = depres_age['Entity'].unique()
arr1=productivity['Country'].unique()


arr2 = []
for name in arr:
    for name2 in arr1:
        if(name == name2):
            arr2.append(name2)
```

```python
productivity.drop(index = [36],inplace=True)

productivity.reset_index(inplace=True)
```

```python
depression_age = depression_age.loc[arr2]
depression_age = depression_age.iloc[:,15:]
depression_age
```

```python
depression_age['Age:15-69_depression_avg']=depression_age.iloc[:,0:14].mean(axis=1,skipna=True)
depression_age.reset_index(inplace=True)

```

```python
productivity = productivity[productivity['Country'].notna()]

a=[]
for i in range(0,61):
    if i%2 != 0:
        a.append(i)

productivity = productivity.drop(a)
productivity.reset_index(inplace = True)

productivity.drop(index = [6,8,17,18,19,21,22,23,24,28,29,30,31],inplace=True)


```

```python
productivity['Average_Productivity']=productivity.iloc[:,2:20].mean(axis=1,skipna=True)
productivity.drop(['level_0','index'],axis=1)

productivity.reset_index(inplace = True)
```

```python
depression_age
```

```python
correl_signif = pearsonr(productivity['Average_Productivity'], depression_age['Age:15-69_depression_avg'])
correl = correl_signif[0]
correl
```

```python
import matplotlib.pyplot as plt

x=productivity.loc[:,'Average_Productivity'].values
y=depression_age.loc[:,'Age:15-69_depression_avg'].values

x = x.reshape(len(productivity), 1)
y = y.reshape(len(depression_age), 1)

from sklearn.linear_model import LinearRegression

regr = LinearRegression()
regr.fit(x, y)
print(regr.score(x, y))


fig, ax = plt.subplots(dpi=100, figsize=(6, 4))
fig.patch.set_facecolor('white')
ax.scatter(x, y,  color='black', label='data')
ax.plot(x, regr.predict(x), color='blue', linewidth=3, label='regression')

ax.set_xlabel('Productivity', fontsize=10)
ax.set_ylabel('Depression', fontsize=10)

ax.legend()
plt.show()
```
